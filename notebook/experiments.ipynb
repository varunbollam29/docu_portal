```python
from dotenv import load_dotenv
import os
from pathlib import Path

env_path = Path().resolve().parent / '.env'
load_dotenv(dotenv_path=env_path)

api_key = os.getenv('API_KEY')
# Strip the extra characters
api_key = api_key.replace('="', '').replace('"', '').strip() if api_key else None

print(f"Cleaned API_KEY: {api_key[:10]}..." if api_key else "None")

from langchain_groq import ChatGroq
llm = ChatGroq(model_name="qwen/qwen3-32b", api_key=api_key)
print(llm.invoke("What is the capital of France?").content)
```

```python
from google import genai
import os
```

```python
embedding_key = os.getenv("embeding_key")  # Note: typo in your .env key name
embedding_key = embedding_key.replace('="', '').replace('"', '').strip() if embedding_key else None
```

```python
client = genai.Client(api_key=embedding_key)
```

```python
result = client.models.embed_content(
    model="gemini-embedding-001",
    contents="What is the meaning of life?"
)

print(result.embeddings)
```
